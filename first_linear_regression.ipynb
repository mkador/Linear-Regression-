```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nasdaq100.csv',sep=';')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df=df.drop('Date',axis = 1) 
```

```python
df.head(3)
```

```python
df
```

```python
plt.scatter(df['Starting (USD)'],df['Ending (USD)'])
plt.title('Stock Prices')
plt.xlabel('Starting (USD)')
plt.ylabel('Ending(USD)')
plt.show()
```

```python
x = df.drop('Ending (USD)',axis = 1)
```

```python
y = df[['Ending (USD)']]
```

```python
#Linear regression
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color = 'red')
plt.scatter(df['Starting (USD)'],df['Ending (USD)'])
plt.title('Stock Prices')
plt.xlabel('Starting (USD)')
plt.ylabel('Ending(USD)')
plt.show()
```

```python
reg.fit(x,y) #train the model
```

```python
#mannually find out result of y
m = reg.coef_
c = reg.intercept_
#y=m*x+c
m*16700+c
```

```python
# using predict function with of linear regression model ,we can pewdict the result of y
reg.predict([[16700]]) # predict for 16700
```

```python
df['Predicted_y'] = reg.predict(x)
df.head(2)
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(x.mean(),y.mean(),color = 'red')
plt.scatter(df['Starting (USD)'],df['Ending (USD)'])
plt.title('Stock Prices')
plt.xlabel('Starting (USD)')
plt.ylabel('Ending(USD)')
plt.show()
```
